## Pedagogical quick-start: the whole story in one pass

If you are reading this cold, keep this chain in your head:

1. **The microscopic state is** `(x, y, d)`: position plus one of four directors.
2. **A noise step** rotates the director but does not translate the walker.
3. **A chiral step** translates first, then rotates; the handedness is controlled by $\omega$.
4. **Small $D_r$** means long deterministic chiral loops, interrupted rarely by noise.
5. **$\omega=0.5$** is the achiral middle point; $\omega=0$ and $\omega=1$ are maximally chiral in opposite senses.
6. **In free space**, chirality suppresses diffusion by trapping the walker in repeated four-step cages.
7. **With open boundaries**, blocked chiral moves create edge localization and boundary currents.
8. **The current split** $J=J_\omega+J_{D_r}$ asks which successful translation followed ordinary chiral motion and which followed a noise event.
9. **The exact solution** is the Perron steady state of the transition matrix; **Fortran MC** is the finite-sampling version of the same dynamics.
10. **The topology enters through spectra**: PBC gives bulk bands, OBC/HPBC reveal edge-localized eigenmodes.

## 1.1 Why active matter on a lattice?

A real chiral active particle &mdash; an *E.&nbsp;coli* with a flagellar bias, a Janus colloid in a
rotating field, an electron in a Hall conductor &mdash; combines three ingredients:
**self-propulsion**, **rotational noise**, and a **chirality bias** (left or right preference
when the noise kicks the orientation).

To study these systems exactly, the authors of Osat *et&nbsp;al.* discretise space and orientation
and reduce the problem to a Markov chain on a 2D lattice. This buys two things:

1. **Closed-form steady state.** The Markov chain has a unique Perron eigenvector $\pi$ that
   is the exact stationary distribution. You can compute it by sparse linear algebra,
   without simulating long trajectories.
2. **Topological structure.** The lattice version exhibits the same edge-state phenomenology
   as continuum chiral active matter, but the operator is a finite-dimensional matrix whose
   spectrum, edge states, and Zak phase you can compute directly.

Discretisation costs us nothing scientifically &mdash; the universal phenomena
(edge localisation, chirality-V in current ratios, gap closure at $\omega=0.5$) all survive.

<div style="background:#eef7ff; border-left:4px solid #3498db; padding:0.8em 1em; margin:1em 0;">
<b>Pedagogical note.</b> Treat the lattice not as an approximation but as the
clean, exactly-solvable arena where you check ideas first, then carry them back to continuum
if you want.
</div>

<a id="sec-1-2"></a>
## 1.2 The TCRW step rule, with a picture

The walker is at site $(x,y)$ with director $d \in \{\uparrow,\rightarrow,\downarrow,\leftarrow\}$.
At each discrete time step exactly one of two branches fires:

**Branch A &mdash; noise step (probability $D_r$).**
The walker stays in place and rotates the director:
$$d \to d_{\rm CCW}\ \text{w.p.}\ \omega,\qquad d \to d_{\rm CW}\ \text{w.p.}\ 1-\omega.$$

**Branch B &mdash; chiral step (probability $1-D_r$).**
The walker translates one site along $d$, **then** rotates the director with the
**opposite** chirality of the noise branch:
$$\begin{aligned}
(x,y) &\to (x+\hat d_x,\,y+\hat d_y) \\
d &\to d_{\rm CW}\ \text{w.p.}\ \omega,\quad d \to d_{\rm CCW}\ \text{w.p.}\ 1-\omega.
\end{aligned}$$

On open boundaries, if the chiral translation would leave the lattice, the walker
**stays at $(x,y)$ and the director is unchanged** &mdash; this is a pure self-loop with
weight $(1-D_r)$. The walker holds its outward-pointing director until the next noise
event flips it. The cumulative dwell at the wall (which scales as $1/D_r$) is the seed of
the topological edge current.

<div style="background:#fff8e1; border-left:4px solid #f1c40f; padding:0.8em 1em; margin:1em 0;">
<b>Sanity check.</b> What happens at $\omega = 1$, $D_r$ small?
<details><summary><i>show answer</i></summary>
Almost every step is chiral (probability $1-D_r \approx 1$) and rotates CW with probability 1.
The walker traces a closed CW 4-cycle:
$(\uparrow) \to (\to) \to (\downarrow) \to (\leftarrow) \to (\uparrow) \cdots$.
Its position oscillates over a $2\!\times\!2$ box. Long-distance diffusion is suppressed, so
$D(\omega\!=\!1)\to 0$ as $D_r\to 0$. This is exactly what paper Fig.&nbsp;1(d) shows.
</details>
</div>

<a id="sec-1-3"></a>
## 1.3 Why split into chiral and noise currents?

Define a flux on each lattice bond from the steady-state Markov chain:
$$J(\mathbf r \to \mathbf r') = \pi(\mathbf r)\, W(\mathbf r' \mid \mathbf r) \;-\; \pi(\mathbf r')\, W(\mathbf r \mid \mathbf r').$$

This $J$ is divergenceless in steady state but is generally **not zero** on a chiral system
&mdash; you have macroscopic loops. Decompose by which branch did the move:

- $J_\omega(\mathbf r \to \mathbf r')$ counts only the moves coming from Branch&nbsp;B (chiral).
- $J_{D_r}(\mathbf r \to \mathbf r')$ counts only the moves coming from Branch&nbsp;A (noise).

Their sum is the full $J$. The **physical meaning**:

- $J_\omega$ is the *direct* topological current: chirality $\omega \neq \tfrac12$ creates a
  net circulation around the boundary.
- $J_{D_r}$ is the *scattering* current: a noise event at the wall bounces the walker
  perpendicular to the wall (this is what panel&nbsp;3(c) and the defect panel&nbsp;2(o) reveal).

Their *ratio* on the wall, $|J_{D_r}|/|J_\omega|$, is the central diagnostic of paper Fig.&nbsp;3(b)(g).
At $\omega = 0.5$ both currents vanish (achiral) but $|J_{D_r}|$ vanishes faster; that's why
the ratio dips to $\sim 10^{-2}$ at the achiral point.

## 2.1 State space and indexing

The walker has $4 (L+1)^2$ states. We pack them into a single integer index using the
**authors' convention**:

$$\boxed{\;s(i,j,d) \;=\; \big(i\,(L+1) + j\big)\,4 + d\;}$$

where $i,j \in \{0,1,\dots,L\}$ and $d \in \{0\!:\!\uparrow, 1\!:\!\to, 2\!:\!\downarrow, 3\!:\!\leftarrow\}$.

Direction-to-vector map:
$$\hat d_0 = (0,+1),\;\hat d_1 = (+1,0),\;\hat d_2 = (0,-1),\;\hat d_3 = (-1,0).$$

CCW rotation maps $d \to (d-1)\bmod 4$; CW maps $d \to (d+1)\bmod 4$.

<a id="sec-2-2"></a>
## 2.2 The transition matrix $W$

$W$ is column-stochastic: column $s$ holds the probabilities $W(s'\!\mid\!s)$ for transitions
*out of* state $s$. Schematically,
$$W = D_r\, R_\omega \;+\; (1-D_r)\, T_\omega \;+\; (1-D_r)\, B_\omega,$$

where:

- $R_\omega$ is the **rotation-only** kernel (Branch&nbsp;A): same site, $d \to d_{\rm CCW}$ w.p.&nbsp;$\omega$,
  $d \to d_{\rm CW}$ w.p.&nbsp;$1-\omega$.
- $T_\omega$ is the **chiral translate-then-rotate** kernel (Branch&nbsp;B), applied to bulk
  sites where the destination is inside the lattice.
- $B_\omega$ is the **boundary self-loop** kernel (Branch&nbsp;B at a wall): stay in place,
  rotate per the same $\omega$ rule.

The split $W = W_\omega + W_{D_r}$ used for the current decomposition has
$W_{D_r} = D_r R_\omega$ and $W_\omega = (1-D_r)(T_\omega + B_\omega)$.

<a id="sec-2-3"></a>
## 2.3 Steady state via Perron eigenvector

For an aperiodic, irreducible chain, $\pi = W\pi$ has a unique solution with $\sum \pi = 1$.
Numerically:

- For small lattices ($L \le 11$): dense `numpy.linalg.eig`.
- For larger ($L \le \sim 100$): sparse shift-invert `scipy.sparse.linalg.eigs(W, k=1, sigma=1)`.
- For very large ($L > 100$): power iteration with explicit sparse mat-vec.

<div style="background:#fdecea; border-left:4px solid #e74c3c; padding:0.8em 1em; margin:1em 0;">
<b>Numerical pitfall (war story #2 below).</b> At $D_r = 1$ the kernel $W = R_\omega$ is
block-diagonal in $(i,j)$ &mdash; every site is its own connected component. The Perron
eigenspace at $\lambda = 1$ becomes $(L+1)^2$-dimensional and the steady state is non-unique.
Cross-checks against the exact $\pi$ at $D_r = 1$ are meaningless. Stop the grid at $D_r = 0.977$.
</div>

<a id="sec-2-4"></a>
## 2.4 The current decomposition

Given $\pi$, define on each oriented bond $(\mathbf r, \mathbf r')$:
$$J_\omega(\mathbf r \to \mathbf r') = \pi_{\mathbf r}\, W_\omega(\mathbf r'\!\mid\!\mathbf r)
   - \pi_{\mathbf r'}\, W_\omega(\mathbf r\!\mid\!\mathbf r'),$$
$$J_{D_r}(\mathbf r \to \mathbf r') = \pi_{\mathbf r}\, W_{D_r}(\mathbf r'\!\mid\!\mathbf r)
   - \pi_{\mathbf r'}\, W_{D_r}(\mathbf r\!\mid\!\mathbf r').$$

These are **per-bond** scalar quantities. To get a **per-site vector**, sum over the four
out-going bonds with their direction vectors:
$$\vec J^{(\alpha)}(\mathbf r) = \sum_{\hat e \in \{\hat x,\hat y,-\hat x,-\hat y\}}
   J^{(\alpha)}(\mathbf r \to \mathbf r + \hat e)\,\hat e.$$

This is exactly what `TRW.calculate_J1_J2_with_boundaries` returns.

<a id="sec-2-5"></a>
## 2.5 Diffusion coefficient via Bloch matrix

For PBC with momentum $\mathbf k$, the transition operator becomes a $4\times 4$ matrix
$P(\mathbf k)$ acting on the director index. Its dominant eigenvalue $\lambda_0(\mathbf k)$
satisfies $\lambda_0(\mathbf 0) = 1$ and, by isotropy of the lattice + the chiral rule,
$$\lambda_0(\mathbf k) = 1 - D\,|\mathbf k|^2 + \mathcal O(|\mathbf k|^4).$$

So $D$ is read off from the Hessian at $\mathbf k = 0$:
$$D = -\tfrac12 \left.\frac{\partial^2 \lambda_0}{\partial k_x^2}\right|_{\mathbf k = 0}
     = -\tfrac12 \left.\frac{\partial^2 \lambda_0}{\partial k_y^2}\right|_{\mathbf k = 0}.$$

This is the "exact" (black-line) reference for paper Fig.&nbsp;1(d), implemented in
`tcrw_fig1_bloch_diffusion.py`. Fortran MC fits the same $D$ from $\langle r^2(t)\rangle / 4t$
at long times.

## 3.1 A minimal Python TCRW simulator

About 25 lines. No optimisations. Use it as a reference implementation to compare against
the optimised production code in `tcrw_fig1_pymc.py`.


````python
import numpy as np

# Direction lookup: 0:up, 1:right, 2:down, 3:left
DX = np.array([ 0, 1, 0,-1])
DY = np.array([ 1, 0,-1, 0])

def tcrw_step(x, y, d, omega, D_r, rng, L=None):
    '''One TCRW step. If L is given, treat as L+1 by L+1 OBC; if None, unbounded bulk.

    Authors' convention:
      - noise step (prob D_r): rotate in place
      - chiral step (prob 1-D_r): translate + rotate (opposite chirality)
      - if chiral translate is blocked by wall: PURE SELF-LOOP, no rotation
    '''
    if rng.random() < D_r:
        # Branch A — noise: rotate only
        if rng.random() < omega:   d = (d - 1) % 4   # CCW
        else:                      d = (d + 1) % 4   # CW
    else:
        # Branch B — chiral translate-then-rotate
        nx, ny = x + DX[d], y + DY[d]
        if L is None or (0 <= nx <= L and 0 <= ny <= L):
            x, y = nx, ny
            if rng.random() < omega:   d = (d + 1) % 4   # CW (opposite chirality)
            else:                      d = (d - 1) % 4   # CCW
        # else: blocked at wall — stay in place, do NOT rotate (authors' convention)
    return x, y, d

# ---- demo: simulate one walker for T steps and plot it ----
import matplotlib.pyplot as plt
rng = np.random.default_rng(seed=1)
omega, D_r, T = 1.0, 1e-3, 5000
xs = np.zeros(T+1, dtype=int); ys = np.zeros(T+1, dtype=int)
x, y, d = 0, 0, 0
for t in range(1, T+1):
    x, y, d = tcrw_step(x, y, d, omega, D_r, rng)
    xs[t], ys[t] = x, y

fig, ax = plt.subplots(figsize=(5.5, 5.5))
ax.plot(xs, ys, lw=0.4, color="0.6")
ax.scatter(xs, ys, c=np.arange(T+1), s=2, cmap="viridis")
ax.set(xlabel="x", ylabel="y", aspect="equal",
       title=f"Minimal TCRW: ω = {omega}, $D_r$ = {D_r}, T = {T}")
ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
````

<div style="background:#eef7ff; border-left:4px solid #3498db; padding:0.8em 1em;">
<b>What you should see.</b> A trajectory that traces tight CW 4-cycles for long stretches
(walker is locked in a chiral orbit) interspersed with rare drifts to a different orbit
when a noise event fires (~1 per $1/D_r = 1000$ steps). Color shows time evolution.
This is the qualitative content of paper Fig.&nbsp;1(b).
</div>

## 3.2 Build $W$ by hand at $L=2$

The smallest non-trivial lattice has $3\times 3 = 9$ sites and $4\times 9 = 36$ states.
Building $W$ explicitly here lets you see the structure.


````python
import numpy as np
import scipy.sparse as sp

def build_W(L, omega, D_r):
    '''Build the sparse transition matrix W for OBC TCRW. Column-stochastic.'''
    n_sites = (L+1)**2
    n = 4 * n_sites
    # use coo_matrix builder
    rows, cols, vals = [], [], []

    def s_idx(i, j, d):
        return (i*(L+1) + j) * 4 + d

    DX = (0, 1, 0,-1); DY = (1, 0,-1, 0)
    for i in range(L+1):
        for j in range(L+1):
            for d in range(4):
                src = s_idx(i, j, d)
                # Branch A: noise rotation (prob D_r)
                d_ccw = (d - 1) % 4
                d_cw  = (d + 1) % 4
                rows += [s_idx(i, j, d_ccw), s_idx(i, j, d_cw)]
                cols += [src, src]
                vals += [D_r * omega, D_r * (1-omega)]
                # Branch B: chiral move (prob 1-D_r) — translate + rotate
                ni, nj = i + DX[d], j + DY[d]
                if 0 <= ni <= L and 0 <= nj <= L:
                    rows += [s_idx(ni, nj, d_cw), s_idx(ni, nj, d_ccw)]
                    cols += [src, src]
                    vals += [(1-D_r) * omega, (1-D_r) * (1-omega)]
                else:
                    # blocked at wall: PURE self-loop, NO rotation (authors' convention)
                    rows += [src]
                    cols += [src]
                    vals += [1 - D_r]
    return sp.csr_matrix((vals, (rows, cols)), shape=(n, n))

W = build_W(L=2, omega=1.0, D_r=0.1)
print(f"W shape: {W.shape}, non-zeros: {W.nnz}")
print(f"Column sums (should all be 1): min={W.sum(axis=0).min():.6f}, "
      f"max={W.sum(axis=0).max():.6f}")
print("\nDensity pattern (each • is a non-zero):")
import sys
dense = W.toarray()
for r in range(W.shape[0]):
    print("  " + "".join("•" if dense[r, c] > 0 else "·" for c in range(W.shape[1])))
````

<div style="background:#eef7ff; border-left:4px solid #3498db; padding:0.8em 1em;">
<b>What to notice.</b> Each column has exactly four non-zeros (two from Branch A, two from
Branch B). Column sums are exactly 1, confirming column-stochastic structure. The pattern
is sparse and structured &mdash; it's why we use sparse linear algebra for production.
</div>

## 3.3 Solve for $\pi$ and inspect it


````python
# Solve W @ pi = pi for the steady-state pi (the right eigenvector at eigenvalue 1)
import scipy.sparse.linalg as spla

W = build_W(L=10, omega=1.0, D_r=1e-3)
# shift-invert for the eigenvalue closest to 1
eigvals, eigvecs = spla.eigs(W, k=1, sigma=1.0, which="LM")
pi = eigvecs[:, 0].real
pi = pi / pi.sum()
print(f"L = 10, ω = 1, D_r = 1e-3")
print(f"  dominant eigenvalue: {eigvals[0].real:.10f}  (should be 1.0)")
print(f"  pi sum: {pi.sum():.6f}, min: {pi.min():.3e}, max: {pi.max():.3e}")

# Marginalise over director to get site occupancy P(x, y)
L = 10
P = pi.reshape((L+1, L+1, 4)).sum(axis=2)
print(f"\nP(x, y) marginal (sum: {P.sum():.6f})")
print(f"  edge sites (4·L = 40): mean P = {(P[0,:].sum() + P[L,:].sum() + P[1:L,0].sum() + P[1:L,L].sum()) / (4*L):.5f}")
print(f"  bulk sites ((L-1)²=81): mean P = {P[1:L, 1:L].mean():.5f}")
print(f"  ratio: {((P[0,:].sum() + P[L,:].sum() + P[1:L,0].sum() + P[1:L,L].sum()) / (4*L)) / P[1:L, 1:L].mean():.1f}")

# Quick visualisation
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
fig, ax = plt.subplots(figsize=(5, 5))
im = ax.imshow(P, origin="lower", norm=mcolors.LogNorm(vmin=1e-5, vmax=1e-1),
               cmap="RdBu_r")
plt.colorbar(im, ax=ax, label="$P(x, y)$")
ax.set(xlabel="x", ylabel="y", title=f"OBC steady-state P(x,y), L={L}, ω=1, $D_r$=$10^{{-3}}$")
plt.tight_layout(); plt.show()
````

## 5.2 Fig.&nbsp;2 &mdash; OBC density, currents, defects

### 5.2.1 Why this figure exists

Once the bulk physics is established (Fig.&nbsp;1), Fig.&nbsp;2 puts the walker inside a *finite
box with hard walls* and asks: **where does the walker spend its time, and what currents
flow in steady state?** The answer is the central topological claim of the paper:

- **Edge localisation** &mdash; the walker spends $\sim 100\!-\!1000\times$ more time at the
  wall than in the bulk (panels a, f, k).
- **Edge current $J_\omega$** &mdash; chirality drives a net circulation around the boundary.
- **Wall-normal noise current $J_{D_r}$** &mdash; a *second* current pointing from the wall
  into the bulk; only visible when something perturbs the wall geometry, e.g. an interior
  L-shape defect (panels m, n, o).

### 5.2.2 Why three columns × three rows in the paper

Panels a&ndash;e: $\omega = 0$ (fully chiral CCW)
Panels f&ndash;j: $\omega = 0.5$ (achiral)
Panels k&ndash;o: $\omega = 0$ + L-shape defect

The L-shape defect (panels k&ndash;o) is the experimental smoking gun: in a clean box the
$J_{D_r}$ signal cancels by symmetry, but as soon as you break the boundary smoothness with
an internal feature, $J_{D_r}$ pops out as a wall-normal arrow pattern.

## 5.4 Fig.&nbsp;4 &mdash; spectrum, edge modes, topology

### 5.4.1 Why this figure exists

After Figs.&nbsp;2 (real-space steady state) and 3 (steady-state observables), Fig.&nbsp;4 takes
a step into **operator theory**: instead of asking "what does the walker do?" it asks
"what is the *spectrum* of the transition operator $W$, and which eigenvectors localise on
the boundary?". This is what makes the model **topological** &mdash; the gap structure
of $W$ in different boundary geometries is the discriminator.

The three boundary geometries used:

- **PBC** (fully periodic, $W$ is block-diagonal in Bloch momentum $\mathbf k$): used to
  establish the bulk gap and identify gap-closing points.
- **OBC** (fully open, $L \times L$ box): used to count edge states.
- **HPBC** (hybrid: y-periodic, x-open): parameterised by $k_y$, used to track edge-state
  dispersion.

### 5.4.8 Notes on convention work

- Each `tcrw_fig4*.py` script independently builds the appropriate transition matrix for
  its boundary condition (PBC / OBC / HPBC), solves the eigenvalue problem with
  `scipy.sparse.linalg.eigs` (or dense `numpy.linalg.eig` at small $L$), and plots the spectrum.
- Edge-state coloring: each eigenvector $|\psi_n\rangle$ is classified by computing
  $\sum_{\text{edge sites}} |\psi_n|^2$ and the rotational sense of the boundary winding.
- The fix to 4(h) caught a sign-convention difference in the Bloch parameterisation
  (translates as $e^{i k_y \hat y}$ vs $e^{-i k_y \hat y}$); after the fix, 4(h) and 4(i)
  agree on the same physics seen from real and imaginary parts respectively.

## 7.1 The L convention drift

**Symptom.** Fortran summary had `L = 10, 19, 49` in column 1, but source said `L_list = (9, 19, 49)`.
Cross-check passed `L = 10` directly to `TRW.build_sparse_transition_matrix(L=10, ...)` &rarr; got
$11\!\times\!11$ grid. Compared to Fortran's $10\!\times\!10$ MC. Off-by-one in $L$ throughout.

**Root cause.** Paper "$L = N$" means **max-index** ($N+1$ sites/side). Fortran source named two
variables, `L_paper` (paper convention) and `L_cur = L_paper + 1` (sites/side, what loops use),
but at one point wrote `L_cur` to the summary instead of `L_paper`.

**Fix.** Three pieces brought back into sync:

1. Fortran source writes `L_paper` to summary &mdash; matches authors' convention.
2. Gnuplot filters on `L_paper` (`==9, ==19, ==49`).
3. Cross-check uses `L_authors = L` (no &minus;1) &mdash; passes L straight to TRW.

After re-running Fortran, all three speak the same language.

<a id="sec-7-2"></a>
## 7.2 The $D_r=1$ singular endpoint

**Symptom.** Last point of panels (a)(b) was either NaN, $10^{15}$ sentinel, or arbitrary.
Cross-check at $D_r = 1$ showed wild residuals.

**Root cause (math).** At $D_r = 1$ no chiral steps fire &mdash; the kernel is pure rotation
$R_\omega$. This is block-diagonal in $(i, j)$: every site is its own connected component.
$\lambda = 1$ has $(L+1)^2$ degenerate eigenvectors, the steady state is non-unique, and
currents are exactly $0/0$.

**Fix.** Stop the $D_r$ grid at $\log_{10}D_r = -0.01$ ($D_r \approx 0.977$).
The divergence of $|J_{D_r}|/|J_\omega|$ at $D_r \to 1$ is still visible without hitting the
singular vertex. Applied in `fig3a.f90`, `fig3b.f90`, `fig3cde.f90`, and the Python `make_fig3`.

<a id="sec-7-3"></a>
## 7.3 Corner-site contamination of wall angles

**Symptom.** Paper Fig.&nbsp;3(j) shows a sharp $\pm\pi/2$ step in $\theta_{J_{D_r}}$ at $\omega = 0.5$.
Our reproduction showed a smooth S-curve through 0 instead.

**Root cause (geometry).** The "left wall" includes corner sites $y = 0$ and $y = L_{\rm cur}-1$,
which **also** belong to the bottom/top walls. Their flux has a *constant* $J_x$ component
(geometry, not chirality) that survives all $\omega$. Including them in $\sum_y J_y / \sum_y J_x$
ruins the angle near $\omega = 0.5$ where the Jy contribution from interior sites is small.

**Fix.** Define a separate **interior-only** sum for $y \in [1, L_{\rm cur}-2]$:

```fortran
! In tcrw_fig3{cde,hij}.f90, write BOTH conventions:
Jx_Dr_in = sum(Jx_Dr(1:L_cur-2))
Jy_Dr_in = sum(Jy_Dr(1:L_cur-2))
! Full-wall (cols 7-8): cross-check vs Fortran-Python
! Interior-only (cols 13-14): paper-style plotting
```

Cross-check stays compatible (full-wall preserved); plot uses interior-only.

<a id="sec-7-4"></a>
## 7.4 Raw counts vs normalised currents

**Symptom.** `tcrw_fig3{cde,hij}_summary.txt` originally wrote integer step counts $J_x, J_y$.
Gnuplot used those raw counts directly for arrow color and length. But $T_{\rm use}$ varies with
$D_r$, so arrows in panel (c) had inconsistent color across the $D_r$ axis.

**Fix.** Reconstruct $T_{\rm use}$ in Fortran (using the same formula `run_one` uses internally)
and write three additional columns: `T_use, |J_Dr|/T_use, |J_om|/T_use`. Gnuplot now reads
the normalised columns; cross-checks unchanged (they normalise themselves).

<a id="sec-7-5"></a>
## 7.5 Quiver length encoding (Python plot)

**Symptom.** `_fig3_plot.py` initially used `arrow_length ∝ log|J|` for **all** quiver panels.
Result: panel (d) at small $D_r$ collapsed to invisible dots; panel (i) lost its rotation
visibility because $|J_\omega|$ has a tiny dynamic range.

**Root cause.** Paper convention is *orientation-only* in (c)(d)(i): direction is the message,
magnitude rides the color. Length-shrinking is a panel-(h) speciality where $|J_{D_r}|$ genuinely
vanishes at $\omega = 0.5$.

**Fix.** Added `length_mode` argument to `_per_y_quiver`:

| Panel | `length_mode` | Behaviour |
|---|---|---|
| (c) | `"fixed"` | every nonzero $J$ gets full arrow length |
| (d) | `"fixed"` | uniform diagonal field across $D_r$ |
| (h) | `"linear_mag"` | length $\propto |J|/|J|_{\max}$, dots at $\omega = 0.5$ |
| (i) | `"fixed"` | rotation visible at every $\omega$ |

Geometry tightened to paper values: `arrow_len_y = 0.40` (was 0.9 = nearly one full site!),
and `arrow_len_x = 0.20` for log-x panels (was $0.50 \times dx_{\rm grid} \approx 0.083$).

## 8.2 DP2 starting points

The codebase is structured so DP2 can extend the dynamics with minimal collateral damage.

### Where to add new physics

| DP2 idea | File to edit | Effort |
|---|---|---|
| Snap (4th-derivative) walker on lattice | new `tcrw_step_snap.f90` + matrix builder | medium |
| Jerky chiral on Tailleur lattice (Stephy Jose direction) | extend `_solve_one` / `build_obc_matrix` in `tcrw_fig3_pymc.py`; reuse `_fig3_plot.py` | small |
| Jerky harmonic oscillator 2D | pure Python (linear ODE), reuses `_fig3_plot.py` | small |
| Collective effects (jerky MIPS) | new Fortran multi-walker driver based on `tcrw_step.f90` | large |
| Jerk in active solid (entropons) | new Bloch-matrix machinery; reuse Fig.&nbsp;4 spectral code | medium |

### Minimal-change pattern for jerky-on-lattice

The cleanest fork point is `tcrw_fig3_pymc.py: build_obc_matrix(...)`:

1. Add a third-order memory state to the index: `s = ((i*(L+1)+j)*4 + d)*M + j_state`.
2. Modify the noise/chiral transition rules so they depend on `j_state`.
3. **All downstream pieces work unchanged**: Perron solve, current decomposition, scan
   helpers, `_fig3_plot.make_figure`.
4. `tcrw_fig3_authors.py` won't work for jerk (TRW.py doesn't know about jerk), but the
   Fortran-MC vs Python-exact cross-check pattern still applies.

### What to write first

Reproduce paper Fig.&nbsp;3 with **jerk added to the noise step rule**. Concretely:

- one new Fortran driver (modify `tcrw_step.f90`),
- one new Python `build_obc_matrix_jerk`,
- one new cross-check.

Result: an honest "jerk on lattice" curve set with the same machinery the paper uses
for non-jerky. This deliverable reuses 90 % of the DP1 codebase.

## 8.3 How to rerun everything

```bash
cd "/path/to/new_fortran_reproduction_and_python"

# 1. Recompile changed Fortran drivers
for n in fig3a fig3b fig3cde fig3hij; do
    gfortran -O2 -fno-range-check -ffree-line-length-none -o tcrw_$n tcrw_$n.f90
    ./tcrw_$n
done

# 2. Render gnuplot PDFs
for f in fig3a fig3b fig3c fig3d fig3e fig3f fig3g fig3h fig3i fig3j; do
    bash run.sh tcrw-plot-${f}-pdf
done

# 3. Render Python figures
python3 tcrw_fig1_pymc.py
python3 tcrw_fig3_authors.py
python3 tcrw_fig3_pymc.py

# 4. Cross-check sanity
for f in fig3a fig3b fig3cde fig3f fig3g fig3hij; do
    python3 tcrw_${f}_crosscheck.py
done
```

**Approximate runtimes** (single CPU, modern laptop):
- `tcrw_fig3a`, `tcrw_fig3b` at $L=49$: ~2 hours each
- `tcrw_fig3cde`, `tcrw_fig3hij` at $L=10$: ~10 min each
- `tcrw_fig3f`, `tcrw_fig3g` at $L=49$: ~1 hour each
- All Python (exact + cross-check + plot): ~2 minutes total
