````python
from __future__ import annotations
from pathlib import Path
import datetime as _dt

ROOT = Path.cwd()

def exists_size(path):
    p = ROOT / path
    if not p.exists():
        return False, None, None
    return True, p.stat().st_size / 1024 / 1024, _dt.datetime.fromtimestamp(p.stat().st_mtime).strftime("%Y-%m-%d %H:%M")

ROOT
````

## What Was Done, Figure By Figure

### Fig. 1

- Reproduced single-walker unwrapped trajectories at $D_r=10^{-3}$ and $T=10^6$.
- Used full step resolution for Fig. 1(b), so four-step cages are visible rather than averaged away.
- Reproduced MSD curves for several $\omega$ values and checked long-time normal diffusion.
- Extracted fitted diffusion coefficients $D(\omega)$ for $D_r=10^{-3}$ and $10^{-2}$.
- Scientific point: chirality suppresses diffusion near $\omega=0$ and $\omega=1$, while diffusion is largest near $\omega=0.5$.

### Fig. 2

- Reproduced OBC clean-box density, trajectories, and currents.
- Corrected the geometry convention to the full $10\times10$ playground, not a smaller wall-ring system.
- Used the authors' boundary-blocking convention through the step kernel.
- Reproduced the current split $J=J_\omega+J_{D_r}$.
- Added/analyzed the L-shape defect geometry.
- Scientific point: the internal defect reveals the wall-normal $J_{D_r}$ component that straight outer-edge symmetry can hide.

### Fig. 3

- Built the Python exact reference from the authors' transition matrix and current decomposition.
- Built Fortran MC drivers and summary files for all Fig. 3 panels.
- Dropped the degenerate $D_r=1$ endpoint.
- Added normalized current columns and interior-only angle columns where needed.
- Aligned Python and gnuplot arrow conventions.
- Created the no-shift cdhi overlay.
- Scientific point: Python exact is the clean reference; Fortran MC explains paper-style rough arrows, especially panel (d) at tiny $D_r$.

### Fig. 4

- Implemented PBC Bloch matrix, OBC transition matrices, and HPBC Fourier blocks.
- Cross-checked OBC matrices element-wise against the authors' original code at small $L$.
- Fixed edge-state coloring using the paper-motivated to-edge / CCW-along-edge classification.
- Built generated panel images and comparison montages.
- Scientific point: the spectral/topological structure is anchored to the same transition matrix convention as the authors' code.
