# TCRW Fig. 1 Artifact Audit

This notebook is a **read-only audit** of what was already reproduced. It does not rerun the Fortran simulations, does not rerun gnuplot, and does not regenerate paper panels.

Goal: document the existing Fig. 1 plots, the source code that produced them, the current data files, and the physics checks behind the reproduction.


```python
from __future__ import annotations
from pathlib import Path
import os
import textwrap
import numpy as np

ROOT = Path.cwd()

def data_rows(path):
    with open(path) as f:
        return sum(1 for line in f if line.strip() and not line.startswith("#"))

def artifact_report(paths):
    print("artifact                                  exists  size_MB  modified")
    for raw in paths:
        p = ROOT / raw
        exists = p.exists()
        size = p.stat().st_size / 1024 / 1024 if exists else float("nan")
        mtime = __import__("datetime").datetime.fromtimestamp(p.stat().st_mtime).strftime("%Y-%m-%d %H:%M") if exists else "-"
        print(f"{raw:42s} {str(exists):>6s} {size:8.3f}  {mtime}")

def source_head(path, n=70):
    p = ROOT / path
    print(f"--- {path} ---")
    with open(p, "r", errors="replace") as f:
        for i, line in zip(range(n), f):
            print(line.rstrip())
```

## What You Did For Fig. 1

1. **Fig. 1(b), trajectories**: wrote `tcrw_fig1b.f90` to generate unwrapped single-walker trajectories for $\omega=0,0.5,0.75,1$ at $D_r=10^{-3}$ and $T=10^6$.
2. Used the paper step rule directly: noise rotates without translation; chiral steps translate first and then rotate.
3. Used unbounded coordinates for the trajectory panel, which is equivalent to PBC-unwrapped motion for a single isolated walker.
4. Wrote every step (`skip=1`) so the tight four-step cages at $\omega=0$ and $\omega=1$ are not lost.
5. **Fig. 1(c), MSD**: wrote `tcrw_fig1c.f90` to compute ensemble MSD on a log-spaced time grid for $\omega=0.5,0.7,0.9,1.0$.
6. **Fig. 1(d), diffusion coefficient**: wrote `tcrw_fig1d.f90` to fit the long-time MSD tail and extract $D(\omega)$ for $D_r=10^{-3}$ and $10^{-2}$.
7. Used `tcrw_fig1b.gnu`, `tcrw_fig1c.gnu`, and `tcrw_fig1d.gnu` to create the final paper-style outputs.
