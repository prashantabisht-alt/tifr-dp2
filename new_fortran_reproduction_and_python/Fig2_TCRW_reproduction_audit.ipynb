## Reader Guide: What Fig. 2 Is Teaching

Fig. 2 is the **real-space boundary physics figure**. It asks what happens when the same walker from Fig. 1 is placed in an open box.

Read the panels in this order:

1. **Occupancy $P(X,Y)$:** the walker spends much more time near walls because chiral attempts into a boundary get blocked.
2. **Trajectory:** the lifted path shows the walker circulating along the boundary rather than wandering uniformly through the box.
3. **Currents:** $J_\omega$ and $J_{D_r}$ are both large near edges, but they partially cancel in $J_{tot}$.
4. **Defect row:** the internal L-shape behaves like a new wall. It creates a second localization ring and exposes the normal component of $J_{D_r}$.

Key physical message: boundaries convert microscopic chirality into macroscopic edge localization and edge currents. The defect panels are not decorative; they reveal current components that a clean straight wall can hide by symmetry.


```python
from __future__ import annotations
from pathlib import Path
import os
import textwrap
import numpy as np

ROOT = Path.cwd()

def data_rows(path):
    with open(path) as f:
        return sum(1 for line in f if line.strip() and not line.startswith("#"))

def artifact_report(paths):
    print("artifact                                  exists  size_MB  modified")
    for raw in paths:
        p = ROOT / raw
        exists = p.exists()
        size = p.stat().st_size / 1024 / 1024 if exists else float("nan")
        mtime = __import__("datetime").datetime.fromtimestamp(p.stat().st_mtime).strftime("%Y-%m-%d %H:%M") if exists else "-"
        print(f"{raw:42s} {str(exists):>6s} {size:8.3f}  {mtime}")

def source_head(path, n=70):
    p = ROOT / path
    print(f"--- {path} ---")
    with open(p, "r", errors="replace") as f:
        for i, line in zip(range(n), f):
            print(line.rstrip())
```

## What You Did For Fig. 2

1. Wrote `tcrw_fig2_clean.f90` for the clean OBC $10\times10$ box.
2. Corrected the convention to a **full $10\times10$ playground**, not a smaller wall-ring interior.
3. Matched the authors' OBC rule: boundary blocking is done by the step kernel bounds check.
4. Ran clean cases for $\omega=0,0.5,1.0$; $\omega=1.0$ is a symmetry check even though it is not a paper row.

```python
print("Row-count checks")
for omega in ["0.0", "0.5", "1.0"]:
    print(f"omega={omega}")
    print("  occ ", data_rows(f"tcrw_fig2_occ_w{omega}.txt"), "expected 100")
    print("  traj", data_rows(f"tcrw_fig2_traj_w{omega}.txt"), "expected 1000000")
    for kind in ["Jtot", "Jomega", "JDr"]:
        print(f"  {kind:6s}", data_rows(f"tcrw_fig2_{kind}_w{omega}.txt"), "expected 100")
print("defect layout", data_rows("tcrw_fig2_defects_layout.txt"), "expected 7")
for kind in ["occ", "traj", "Jtot", "Jomega", "JDr"]:
    file = "tcrw_fig2_occ_defects.txt" if kind == "occ" else "tcrw_fig2_traj_defects.txt" if kind == "traj" else f"tcrw_fig2_{kind}_defects.txt"
    print(f"defect {kind:6s}", data_rows(file))
```

```python
print("Read-only density and current consistency checks")
for omega in ["0.0", "0.5", "1.0"]:
    occ = np.loadtxt(f"tcrw_fig2_occ_w{omega}.txt", comments="#")
    print(f"omega={omega}: sum P={occ[:,2].sum():.12f}, max P={occ[:,2].max():.5e}")
occ_def = np.loadtxt("tcrw_fig2_occ_defects.txt", comments="#")
print(f"defect: sum P={occ_def[:,2].sum():.12f}, max P={occ_def[:,2].max():.5e}")

def current_residual(suffix):
    jt = np.loadtxt(f"tcrw_fig2_Jtot_{suffix}.txt", comments="#")
    jo = np.loadtxt(f"tcrw_fig2_Jomega_{suffix}.txt", comments="#")
    jd = np.loadtxt(f"tcrw_fig2_JDr_{suffix}.txt", comments="#")
    # files are written in same coordinate order
    res = np.linalg.norm(jt[:,2:4] - jo[:,2:4] - jd[:,2:4], axis=1)
    return res.max()
for suffix in ["w0.0", "w0.5", "w1.0", "defects"]:
    print(f"{suffix:8s}: max |Jtot-Jomega-JDr| = {current_residual(suffix):.3e}")
```
